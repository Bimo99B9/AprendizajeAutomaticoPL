# Sesión 4 - Representación gráfica

En esta sesión práctica vamos a ver cómo podemos realizar representaciones gráficas básicas. Para ello vamos a utilizar la librería `matplotlib` que tiene muchas más posibilidades de las que vamos a mostrar aquí. Existen más librerías para la representación gráfica, pero hoy en día `matplotlib` es la más popular.

https://matplotlib.org/stable/index.html

Para utilizar esta librería es necesario importarla. Habitualmente se hace de la siguiente manera:

```python
import matplotlib.pyplot as plt
```

Podéis ver todas las funcionalidades que tiene en:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html?highlight=plot#module-matplotlib.pyplot

## 4.1 Primer gráfico
Vamos a comenzar representando un gráfico básico:

```python
# representamos un gráfico sencillo
plt.plot([1, 2, 3, 4], [1, 4, 2, 3])  
plt.show()
```

Existe una figura por defecto a la que podemos acceder a través del nombre `plt`.

Hemos utilizado la función `plot()` y le hemos pasado un vector con los valores de $x$ y otro con los valores de $y$.

Finalmente hemos mostrado la figura utilizando la función `show()`.

Lo que estamos viendo, incluyendo todas las zonas blancas, se conoce como **figura**.

Dentro de la figura podemos crear gráficos (en este caso tenemos sólo uno).


## 4.2 Borrado
Si queremos podemos borrar el gráfico que hemos representado utlizando `cla()`.

```python
plt.cla()
plt.show()
```

Y también podemos borrar la figura al completo con `clf()`:

```python
plt.clf()
plt.show()
```

Esta es una manera muy sencilla de crear gráficos y veréis que es extensamente utilizada.

## 4.3 Cómo crear figuras

Hay varias maneras de crear figuras.

Podemos crear una figura vacía:

```python
fig = plt.figure()
plt.show()
```

Podemos crear una figura con espacio para un gráfico:

```python
fig, ax = plt.subplots()
plt.show()
```

O podemos, finalmente, crear una figura con espacio para los gráficos que indiquemos:

```python
fig, axs = plt.subplots(4, 3)
plt.show()
```

Podemos cambiar el tamaño de la figura utilizando la función `set_size_inches()`, que debe recibir como parámetro una tupla: `(ancho, alto)`: 

```python
fig, axs = plt.subplots(4, 3)
fig.set_size_inches((12,15))
plt.show()
```

Vamos a crear un conjunto de datos y vamos a utilizarlo para ir mostrando diferentes tipos de gráficos:

```python
from sklearn import datasets
X, y = datasets.make_classification(n_samples=200, n_features=5, n_informative=3, n_redundant=1, n_classes=2)

print("Matriz de datos:\n", X[:4])
print("Clase:\n", y)
```

## 4.4 Aspecto de los gráficos

Vamos a representar los valores contenidos en el atributo de la columna 0 con los valores del atributo de la columna 1. Para ello utilizaremos la función `scatter()` que nos muestra un gráfico de dispersión.

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los datos en el gráfico
ax.scatter(x=X[:,0], y=X[:,1])

plt.show()
```

### 4.4.1 Colores
Si queremos podemos utilizar diferentes colores para representar puntos de series diferentes. Los colores disponibles están en https://matplotlib.org/stable/gallery/color/named_colors.html


```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b')

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r')

plt.show()
```

### 4.4.2 Título
Podemos ponerle título utilizando `set_title()`:

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b')

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r')

# ponemos título
ax.set_title('Atributo 0 vs. atributo 1 en función de la clase')

plt.show()
```

### 4.4.3 Ejes
Poner nombre a los ejes con `set_xlabel()` y `set_ylabel()`: 

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b')

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r')

# ponemos título
ax.set_title('Atributo 0 vs. atributo 1 en función de la clase')

# ponemos nombre a los ejes
ax.set_xlabel('Atributo 0')
ax.set_ylabel('Atributo 1') 

plt.show()
```

Y también podemos modificar los ejes a nuestra conveniencia utilizando:
- `set_xticks()` y `set_yticks()` para indicar qué valores queremos ver en los ejes 
- `set_xlim()` y `set_ylim()` para indicar valores mínimos y máximos de cada eje 
- `set_xscale()` y `set_yscale()` para establecer la escala de cada eje, que puede ser lineal (`linear`) o logarítmica (`log`) 

### 4.4.4 Leyenda

Es común también incorporar leyendas a nuestros gráficos. Para poder hacerlo, cada vez que incorporemos una serie de datos al gráfico **debemos asignarle una etiqueta**, que será el texto que aparecerá en la leyenda.

Para mostrar la leyenda debemos utilizar `legend()` indicando el lugar en el que deseamos que aparezca. Podemos dejar que `matplotlib` elija la mejor ubicación (`loc='best'`) o podemos indicar nosotros la ubicación deseada (`'upper left', 'upper right', 'lower left', 'lower right', 'upper center', 'lower center', 'center left', 'center right'`):

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b', label='clase 0') # añadimos etiqueta

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r', label='clase 1') # añadimos etiqueta

# ponemos título
ax.set_title('Atributo 0 vs. atributo 1 en función de la clase')

# ponemos nombre a los ejes
ax.set_xlabel('Atributo 0')
ax.set_ylabel('Atributo 1') 

# incorporamos la leyenda
ax.legend(loc='best')

plt.show()
```

### 4.4.5 Rejilla

Utilizando la función `grid()` podemos incorporar una rejilla a nuestro gráfico. Tiene múltiples opciones que pueden consultarse en: https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html?highlight=grid#matplotlib.axes.Axes.grid 

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b', label='clase 0') 

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r', label='clase 1') 

# ponemos título
ax.set_title('Atributo 0 vs. atributo 1 en función de la clase')

# ponemos nombre a los ejes
ax.set_xlabel('Atributo 0')
ax.set_ylabel('Atributo 1') 

# incorporamos la leyenda
ax.legend(loc='best')

# incorporamos una rejilla
ax.grid(axis='both', color='gray', linestyle='--')

plt.show()
```

### 4.4.6 Marcadores

También podemos cambiar la forma de los marcadores indicándolo en el momento en el que se incorpora la serie de datos al gráfico, en este caso cuando utilizamos `scatter()`. Podemos consultar los marcadores disponibles en: https://matplotlib.org/stable/api/markers_api.html

```python
# creamos una figura con un solo gráfico
fig, ax = plt.subplots()

# representamos los ejemplos de la clase 0 en azul
ejem_clase_0 = y == 0
ax.scatter(x=X[ejem_clase_0, 0], y=X[ejem_clase_0, 1], color='b', label='clase 0', marker='*') 

# representamos los ejemplos de la clase 1 en rojo
ejem_clase_1 = y == 1
ax.scatter(x=X[ejem_clase_1, 0], y=X[ejem_clase_1, 1], color='r', label='clase 1', marker='X') 

# ponemos título
ax.set_title('Atributo 0 vs. atributo 1 en función de la clase')

# ponemos nombre a los ejes
ax.set_xlabel('Atributo 0')
ax.set_ylabel('Atributo 1') 

# incorporamos la leyenda
ax.legend(loc='best')

# incorporamos una rejilla
ax.grid(axis='both', color='gray', linestyle='--')

plt.show()
```

## 4.5 Varios gráficos en una figura

Hemos estado trabajando hasta ahora únicamente con dos de los cinco atributos que tiene el conjunto que hemos creado. 

¿Por qué no hacer una figura que contenga todos los gráficos posibles comparando los atrubutos dos a dos?

Lo que tendremos que hacer es crear una figura en la que haya espacio para los 25 gráficos que necesitamos y después ir representando lo que queramos en cada gráfico:

```python
num_atributos = X.shape[1]

fig, axs = plt.subplots(num_atributos, num_atributos)
fig.set_size_inches((num_atributos*3, num_atributos*3))

ejem_clase_0 = y == 0
ejem_clase_1 = y == 1

for fila in range(num_atributos):
    for columna in range(num_atributos):
        axs[fila][columna].scatter(x=X[ejem_clase_0, columna], y=X[ejem_clase_0, fila], color='b', label='clase 0') 
        axs[fila][columna].scatter(x=X[ejem_clase_1, columna], y=X[ejem_clase_1, fila], color='r', label='clase 1') 

        if columna == 0:
            axs[fila][columna].set_ylabel('Atributo ' + str(fila)) 
        if fila == num_atributos-1:
            axs[fila][columna].set_xlabel('Atributo ' + str(columna))
        
        if fila == 0 and columna == 0:
            axs[fila][columna].legend(loc='best')

        axs[fila][columna].grid(axis='both', color='gray', linestyle='--')

plt.show()
```

## 4.6 Guardar un gráfico en fichero

Si queremos podemos guardar la figura en un fichero utilizando la función `savefig()`.

Por defecto guarda las figuras en formato PNG, pero podemos guardarlas en otro formato utilizando el parámetro `format`. Tiene otras muchas posibilidades que pueden consultarse en: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html?highlight=savefig#matplotlib.pyplot.savefig


```python
num_atributos = X.shape[1]

fig, axs = plt.subplots(num_atributos, num_atributos)
fig.set_size_inches((num_atributos*3, num_atributos*3))

ejem_clase_0 = y == 0
ejem_clase_1 = y == 1

for fila in range(num_atributos):
    for columna in range(num_atributos):
        axs[fila][columna].scatter(x=X[ejem_clase_0, columna], y=X[ejem_clase_0, fila], color='b', label='clase 0') 
        axs[fila][columna].scatter(x=X[ejem_clase_1, columna], y=X[ejem_clase_1, fila], color='r', label='clase 1') 

        if columna == 0:
            axs[fila][columna].set_ylabel('Atributo ' + str(fila)) 
        if fila == num_atributos-1:
            axs[fila][columna].set_xlabel('Atributo ' + str(columna))
        
        if fila == 0 and columna == 0:
            axs[fila][columna].legend(loc='best')

        axs[fila][columna].grid(axis='both', color='gray', linestyle='--')

plt.savefig('scatter_de_atributos.jpg', format='jpg')
```

## 4.7 Otros tipos de gráfico

Vamos a ver algunos tipos de gráficos.

### 4.7.1 Diagrama de barras

Podemos utilizar `bar()` para hacer diagramas de barras verticales:

```python
fig, ax = plt.subplots()

ax.bar(['clase 0', 'clase 1'], [(y==0).sum(), (y==1).sum()])

plt.show()
```

Como hay 100 ejemplos de cada clase las barras tienen la misma altura.

También podemos utilizar `barh()` para hacer un diagrama de barras de manera horizontal:


```python
fig, ax = plt.subplots()

ax.barh(['Atributo 0', 'Atributo 1', 'Atributo 2', 'Atributo 3', 'Atributo 4'], 
[(X[:,0]>0).sum(), (X[:,1]>0).sum(), (X[:,2]>0).sum(), (X[:,3]>0).sum(), (X[:,4]>0).sum()])

ax.set_title('Valores mayores que 0')

plt.show()
```

### 4.7.2 Histogramas

Con `hist()` podemos hacer histogramas:

```python
fig, ax = plt.subplots()

ax.hist(X[:,0], 10)

ax.set_title('Distribución del atributo 0')

plt.show()
```

### 4.7.3 Diagrama de líneas

Ya vimos al principio que se puede prepresentar utilizando la función `plot()` pasándole los valores de las abscisas (x) y de las ordenadas (y).

Cuando le damos un sólo vector, considera que ese vector son las ordenas y para las abscisas crea un vector de de valores crecientes (0..N-1).

Vamos a ver un ejemplo de utilización:

```python
import numpy as np

fig, ax = plt.subplots()

for i in range(X.shape[1]):
    ax.plot(np.sort(X[:,i]), label='Atributo '+str(i))

ax.legend(loc='best')
ax.set_title('Valores de los atributos')

plt.show()
```
